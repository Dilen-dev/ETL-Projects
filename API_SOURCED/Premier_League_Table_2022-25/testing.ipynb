## TRANSFORM


```python
payload
```

```python
standings_list = payload['response'][0]['league']['standings'][0]
standings_list
```

```python
formatted_standings_list = json.dumps(standings_list, indent=2)
print(formatted_standings_list)
```

```python
rows = []
column_names = ['season','position','team_id','team','played','won','draw','lost','goals_for','goals_against','goal_diff','points','form',]

for club in standings_list:
    season = 2023
    position = club['rank']
    team_id = club['team']['id']
    team = club['team']['name']
    played = club['all']['played']
    won = club['all']['win']
    draw = club['all']['draw']
    lost = club['all']['lose']
    goals_for = club['all']['goals']['for']
    goals_against = club['all']['goals']['against']
    goals_diff = club['goalsDiff']
    points = club['points']
    form = club['form']

    tuple_for_clubs = (season,position,team_id,team,played,won,draw,lost,goals_for,goals_against,goals_diff,points,form)

    rows.append(tuple_for_clubs)
```

```python
rows
```

```python
df = pd.DataFrame(rows, columns=column_names)
df.head(20)
```

## LOAD

```python
MYSQL_HOST = os.getenv("MYSQL_HOST")
MYSQL_USER = os.getenv('MYSQL_USER')
MYSQL_PASSWORD = os.getenv('MYSQL_PASSWORD')
MYSQL_PORT = os.getenv('MYSQL_PORT')
MYSQL_DATABASE = os.getenv('MYSQL_DATABASE')
```

```python
conn = connector.connect(
    host = MYSQL_HOST,
    port = MYSQL_PORT,
    user = MYSQL_USER,
    password = MYSQL_PASSWORD,
    connection_timeout = 10,
    autocommit = False,
    raise_on_warnings = True
)
server_cur = conn.cursor()
print(f"[SUCCESS] - Connected to MySQL server")
```

```python
server_cur.close()
conn.close()
```

```python
db_conn = connector.connect(
    host = MYSQL_HOST,
    port = MYSQL_PORT,
    user = MYSQL_USER,
    password = MYSQL_PASSWORD,
    database = MYSQL_DATABASE
)

cur = db_conn.cursor()
print(f"[SUCCESS] - Connected to MySQL database")
```

```python
table= 'standings'
cur.execute('show tables like %s',(f"{table}",))

if cur.fetchone() is None:
    raise SystemExit(f"This table {table} does not exist, please create it...")
else:
    print(f"[SUCCESS] - This {table} exists!!! Continue to next task..")
```

```python
# UPSERT Operation

table_cols = ['season','position','team_id','team','played','won','draw','lost','goals_for','goals_against','goal_diff','points','form',]

standings_df = df[table_cols]
standings_df
```

```python
standings_records_tuple = standings_df.itertuples(index = False, name = None)

list_of_standings_records_tuples = list(standings_records_tuple)
```

```python
UPSERT_SQL = f""" 
INSERT INTO {table}
(season,position,team_id,team,played,won,draw,lost,goals_for,goals_against,goal_diff,points,form)
VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s) AS src
ON DUPLICATE KEY UPDATE
position = src.position,
team = src.team,
played = src.played,
won = src.won,
draw = src.draw,
lost = src.lost,
goals_for = src.goals_for,
goals_against = src.goals_against,
goal_diff = src.goal_diff,
points = src.points,
form = src.form;
"""
```

```python
no_of_rows_uploaded_to_mysql = len(list_of_standings_records_tuples)
```

```python
try:
    cur.executemany(UPSERT_SQL, list_of_standings_records_tuples)
    db_conn.commit()
    print(f"[SUCCESS] - Upsert attempted for {no_of_rows_uploaded_to_mysql} rows!")
except Exception as e:
    db_conn.rollback()
    print(f"[ERROR] - Rolled back due to this... {e}")
finally:
    cur.close()
    db_conn.close()
    print("All connections have been closed. \n\nThe Clean up is completed!!")
```
